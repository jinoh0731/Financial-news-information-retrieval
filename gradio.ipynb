```python
import pandas as pd
import gradio as gr
```

```python
df_ada = pd.read_csv('ADA/final_openai_dataset_for_demo.csv')
```

```python
# old function without URL link

# Function to filter, sort, and select specific data based on user inputs
def get_articles_no_url(industry, num_articles):
    # Filter data based on the selected industry
    filtered_data = df_ada[df_ada['Industry'] == industry]
    # Sort the data by 'similarity_score' in descending order
    sorted_data = filtered_data.sort_values(by='max_cosine_similarities', ascending=False)

    selected_data = sorted_data[['Company', 'Ticker', 'title', 'url']].head(num_articles)

    
    return selected_data
```

```python
def get_articles(industry, num_articles):
    # Filter and sort data
    filtered_data = df_ada[df_ada['Industry'] == industry]
    sorted_data = filtered_data.sort_values(by='max_cosine_similarities', ascending=False)
    selected_data = sorted_data[['Company', 'Ticker', 'title', 'url']].head(num_articles)

    # Create HTML links for display
    links_html = "<ul>"
    for _, row in selected_data.iterrows():
        links_html += f"<li><a href='{row['url']}' target='_blank'>{row['title']}</a> - {row['Company']} ({row['Ticker']})</li>"
    links_html += "</ul>"

    return links_html
```

```python
# Create a dropdown for Industry selection and an input field for the number of articles
industry_dropdown = gr.Dropdown(choices=sorted(df_ada['Industry'].unique()), label="Select Industry")
number_input = gr.Number(value=3, label="Number of Articles")
```

```python
import gradio as gr

with gr.Blocks() as app:
    with gr.Row():
        gr.Markdown("# ESG News Retrieval System")
        industry_dropdown = gr.Dropdown(label="Select Industry", choices=sorted(df_ada['Industry'].unique()))
        number_input = gr.Number(label="Number of Articles", value=3)
        submit_button = gr.Button("Submit")
    with gr.Row():
        output = gr.HTML()  # Use HTML component to render clickable links

    submit_button.click(
        fn=get_articles,
        inputs=[industry_dropdown, number_input],
        outputs=output
    )

app.launch()
```
